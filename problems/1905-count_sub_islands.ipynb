```python
def prompt(value = None):
    import numpy as np

    if not isinstance(value, str) and not isinstance(value, list):
        value = input()
        
    try:
        return np.matrix(eval(value))
    except:
        return np.matrix(value)

def draw(matrix):
    import matplotlib.pyplot as plt

    _, ax = plt.subplots()

    ax.matshow(matrix, cmap='Pastel1_r', interpolation='none')

    I, J = matrix.shape

    print(f"matrix.shape({I},{J})")

    for i in range(I):
        for j in range(J):
            ax.text(j, i, str(matrix[i, j]), va='center', ha='center')
```

```python
draw(prompt())
```
